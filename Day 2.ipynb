# Modules

Python has a very rich module library that has several functions to do many tasks. 
‘import’ keyword is used to import a particular module into your python code. For instance consider the following program.

```python
# math module 
import math 
  
def Main(): 
    num = -60
  
    # fabs is used to get the absolute  
    # value of a decimal 
    num = math.fabs(num)  
    print(num) 
      
      
if __name__=="__main__": 
    Main() 
```

# Displaying the specific rows using “iloc” and “loc” functions.

```python
data.iloc[5] 
#it will display records only with species "Iris-setosa". 
data.loc[data["species"] == "setosa"] 
```

```python
data.iloc[11] 
#it will display records only with species "virginica". 
data.loc[data["species"] == "virginica"] 
```

# Counting the number of counts of unique values using “value_counts()”.

```python
#In this dataset we will work on the Species column, it will count number of times a particular species has occurred. 
data["species"].value_counts() 
#it will display in descending order. 
```

# Calculating sum, mean and mode of a particular column.

```python
sum_data = data["sepal_length"].sum() 
mean_data = data["sepal_length"].mean() 
median_data = data["sepal_length"].median() 
  
print("Sum:",sum_data, "\nMean:", mean_data, "\nMedian:",median_data) 
```

# Extracting minimum and maximum from a column.

```python
min_data=data["sepal_length"].min() 
max_data=data["sepal_length"].max() 
  
print("Minimum:",min_data, "\nMaximum:", max_data) 
```
